Suppose there is an apparel company that coordinates multiple stores. You as an employee of this company, checks the order management table on a daily basis for each supplier companies.

```python
import pandas as pd
import openpyxl 
import glob 

#the library glob can get the file name that matches a specific condition
```

# Case 1: how to split in several excel files

```python
#Define variables for the excel file path, sheet name and the path where the splitted files will be saved 
import_file_path= 'sample-1.xlsx' #input the file path where the file is saved
#e.g: import_file_path='/Users/kaggle/Documents/sample-1.xlsx'
excel_sheet_name='Order'
# export_file_path= '/Users/kaggle/Documents/folder' input the folder path that the splitted files will be saved  

# by doing so, it is easy to change only here when using it on another file/folders. 
```

```python
df_order= pd.read_excel(import_file_path, sheet_name= excel_sheet_name)
df_order
```

```python
#Define a DataFrame with all supplier`s names with no duplicates in 'Company name' column
company_name= df_order['Company name'].unique()
company_name 
```

```python
#Splitting the dataframe by each company name
#Firstly, let`s save the one that matches only the lines related to Company A
df_order[df_order['Company name']== 'Company A']
```

```python
# Let`s split the data for each company using the for function. 
for i in company_name:
    df_order_company= df_order[df_order['Company name']==i]
    print(df_order_company)
```

```python
#Saving each company in excel file separetely 
for i in company_name:
    df_order_company= df_order[df_order['Company name']==i]
    #Combine the filename and xlsx extension
    df_order_company.to_excel(i+'.xlsx')
    #df_order_company.to_excel(export_file_path+'/'+i+'.xlsx')
```

The file has been created on the export file path.

# Case 2: How to concatenate several excel files in one file

```python
#Define variables for the excel file path, sheet name and the path where the concatenated file will be saved
#export_file_path2= '/Users/kaggle/Documents/Concat'

#import_folder_path= '/Users/kaggle/Documents/Split'

path='*.xlsx'
#path=import_folder_path+'/'+'*.xlsx' 
#Multiply * takes all files with xlsx extension existing in the import folder path
```

```python
#Using glob we can get all the xlsx files that was separeted in the previous process inside the 'path'
file_path=glob.glob(path)
file_path
```

```python
#read all excel files and save in df_concat variable
df_concat=pd.DataFrame() #create a empty data frame

#the file names aquired in i will be read and saved in data Frame
for i in file_path:
    df_read_excel=pd.read_excel(i)
    print(df_read_excel.head(3))
```

```python
#Concatenating all those excel files in one dataframe
df_concat=pd.DataFrame()

for i in file_path:
    df_read_excel=pd.read_excel(i)
    df_concat=pd.concat([df_read_excel, df_concat])
```

```python
df_concat
```

```python
df_drop= df_concat.drop('Unnamed: 0', axis=1)
df_drop.head(3)
```

```python
#sort the data by Payment order column in descending order
df_sort=df_drop.sort_values(by='Payment order', ascending=False)
df_sort
```

```python
#save the dataframe to excel file
df_sort.to_excel('Payment_order.xlsx') 
#df_sort.to_excel('export_file_path2+'/'+'/Payment_order.xlsx') 
```

```python
workbook= openpyxl.load_workbook('Payment_order.xlsx') #open the excel file
#workbook= openpyxl.load_workbook(export_file_path2+'/Payment_order.xlsx')
worksheet= workbook.worksheets[0] #select the excel sheet
worksheet.delete_cols(1) #delete the index column
workbook.save('Payment_order.xlsx') #save the file and it is ready to be sent by e-mail
#workbook.save(export_file_path2+'/Payment_order.xlsx')
```
